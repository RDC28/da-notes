```python
# lambda function

def square(num):
    return num ** 2
```

```python
square(2)
```

```python
list(map(lambda x: x * 2, [1, 2, 3, 4, 5]))
```

```python
list(map(lambda x: [i * 2 for i in x], [[1, 2, 3], [4, 5, 6]]))

filter()
```

```python
# filter

list(filter(lambda x: int(x) % 2 == 1 if x != '0' else False, '0123456789'))
```

```python
from functools import reduce

reduce(lambda x, y: x + y, [1, 2, 3, 4])
```

```python
# map

list(map(lambda x: (x[0] ** 2, x[1] ** 2) if x[0] == x[1] else (x[0], x[1]), {1: 2, 2: 2}.items()))
```

```python
list(map(lambda x : x + 2 , [1, 2, 3, 4, 5]))

list(map(lambda x: chr(ord(x) + 2), 'rohit'))
```

```python
list(filter(lambda x: ord(x) >= 65 and ord(x) < 97, 'ABCDabcdEFGH'))
```

```python
reduce(lambda x, y: x * y, range(1, 5 + 1))
```
